### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
print("Datos de la tienda 1".center(60,"*"),"\n",tienda,"\n")
print("Datos de la tienda 2".center(60,"*"),"\n",tienda2,"\n")
print("Datos de la tienda 3".center(60,"*"),"\n",tienda3,"\n")
print("Datos de la tienda 4".center(60,"*"),"\n",tienda4,"\n")
```

```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingreso total Tienda 1: $", ingreso_tienda1)
print("Ingreso total Tienda 2: $", ingreso_tienda2)
print("Ingreso total Tienda 3: $", ingreso_tienda3)
print("Ingreso total Tienda 4: $", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
print(tienda.columns)

```

```python
ventas_por_categoria_tienda1 = tienda["Categoría del Producto"].value_counts()
ventas_por_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventas_por_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventas_por_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

print("Ventas por categoría Tienda 1: \n", ventas_por_categoria_tienda1, "\n")
print("Ventas por categoría Tienda 2: \n", ventas_por_categoria_tienda2, "\n")
print("Ventas por categoría Tienda 3: \n", ventas_por_categoria_tienda3, "\n")
print("Ventas por categoría Tienda 4: \n", ventas_por_categoria_tienda4, "\n")
```

```python
import matplotlib.pyplot as plt

ventas_por_categoria_tienda1.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de Productos Vendidos")
plt.show()

ventas_por_categoria_tienda2.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 2")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de Productos Vendidos")
plt.show()

ventas_por_categoria_tienda3.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 3")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de Productos Vendidos")
plt.show()

ventas_por_categoria_tienda4.plot(kind="bar")
plt.title("Ventas por Categoría - Tienda 4")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de Productos Vendidos")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = tienda["Calificación"].mean()
promedio_calificacion_tienda2 = tienda2["Calificación"].mean()
promedio_calificacion_tienda3 = tienda3["Calificación"].mean()
promedio_calificacion_tienda4 = tienda4["Calificación"].mean()

print("Calificación promedio Tienda 1:", round(promedio_calificacion_tienda1,2))
print("Calificación promedio Tienda 2:", round(promedio_calificacion_tienda2,2))
print("Calificación promedio Tienda 3:", round(promedio_calificacion_tienda3,2))
print("Calificación promedio Tienda 4:", round(promedio_calificacion_tienda4,2))
```

```python
import matplotlib.pyplot as plt

promedios = [promedio_calificacion_tienda1, promedio_calificacion_tienda2, promedio_calificacion_tienda3, promedio_calificacion_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, promedios, color="skyblue")
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)  # suponiendo que la escala es de 0 a 5
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos_tienda1 = tienda["Producto"].value_counts()
productos_mas_vendidos_tienda2 = tienda2["Producto"].value_counts()
productos_mas_vendidos_tienda3 = tienda3["Producto"].value_counts()
productos_mas_vendidos_tienda4 = tienda4["Producto"].value_counts()

print("Más vendido Tienda 1:")
print(productos_mas_vendidos_tienda1.head(1),"\n")
print("Menos vendido Tienda 1:")
print(productos_mas_vendidos_tienda1.tail(1),"\n\n")

print("Más vendido Tienda 2:")
print(productos_mas_vendidos_tienda2.head(1),"\n")
print("Menos vendido Tienda 2:")
print(productos_mas_vendidos_tienda2.tail(1),"\n\n")

print("Más vendido Tienda 3:")
print(productos_mas_vendidos_tienda3.head(1),"\n")
print("Menos vendido Tienda 3:")
print(productos_mas_vendidos_tienda3.tail(1),"\n\n")

print("Más vendido Tienda 4:")
print(productos_mas_vendidos_tienda4.head(1),"\n")
print("Menos vendido Tienda 4:")
print(productos_mas_vendidos_tienda4.tail(1),"\n\n")
```

```python
import matplotlib.pyplot as plt

productos_mas_vendidos_tienda1.head(5).plot(kind="bar")
plt.title("Top 5 Productos Más Vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.show()

productos_mas_vendidos_tienda2.head(5).plot(kind="bar")
plt.title("Top 5 Productos Más Vendidos - Tienda 2")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.show()

productos_mas_vendidos_tienda3.head(5).plot(kind="bar")
plt.title("Top 5 Productos Más Vendidos - Tienda 3")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.show()

productos_mas_vendidos_tienda4.head(5).plot(kind="bar")
plt.title("Top 5 Productos Más Vendidos - Tienda 4")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1 = tienda["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio Tienda 1: $", round(envio_promedio_tienda1,2))
print("Costo de envío promedio Tienda 2: $", round(envio_promedio_tienda2,2))
print("Costo de envío promedio Tienda 3: $", round(envio_promedio_tienda3,2))
print("Costo de envío promedio Tienda 4: $", round(envio_promedio_tienda4,2))
```

```python
import matplotlib.pyplot as plt

envios = [envio_promedio_tienda1, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, envios, color="orange")
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío Promedio ($)")
plt.show()
```
